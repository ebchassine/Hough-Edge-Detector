```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = [(10, 10), (20, 20), (30, 30)]

theta = np.linspace(0, np.pi, 1000) 

# Compute rho for each point
curves = []
for x, y in points:
    rho = x * np.cos(theta) + y * np.sin(theta)
    curves.append(rho)

# Plot sinusoids in Hough space
plt.figure(figsize=(10, 6))
for i, rho in enumerate(curves):
    plt.plot(theta, rho, label=f"Point ({points[i][0]}, {points[i][1]})")

plt.title("Sinusoids in Hough Space")
plt.xlabel(r"$\theta$ (radians)")
plt.ylabel(r"$\rho$")
plt.legend()
plt.grid()
plt.show()

rho_intersect = np.intersect1d(np.round(curves[0], 1), np.round(curves[1], 1))[0]
theta_intersect = theta[np.argmin(np.abs(curves[0] - rho_intersect))]

# Convert from (rho, theta) to (m, c) in y = mx + c
m = -np.cos(theta_intersect) / np.sin(theta_intersect)
c = rho_intersect / np.sin(theta_intersect)
print(f"(m, c) are {m.round(2)},{c.round(2)}")
```
